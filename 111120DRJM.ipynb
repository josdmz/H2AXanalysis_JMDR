```python

rows = []

for image in range(1,136):
    
    
    df_slice = df.loc [ df['ImageNumber'] == image].reset_index()

    #image 1 has 85 objects
    objects = len(df_imageNumbers.get_group(image)['ObjectNumber'])
    
    group = list(df_slice['Group'])[0]

  
    # Prints list of random items of given length 
    list1 = list(range(0,objects))

    sampling = sample(list1,15)

    #print(sampling)

    #df_1
    sampling_rows = df_slice.iloc[ sampling,:]

    h2ax_sum = sampling_rows['Parent_h2AX'].sum()
    h2ax_mean = sampling_rows['Parent_h2AX'].mean()
    positive = len(sampling_rows.loc[sampling_rows['Parent_h2AX'] > 0] )

    #print(h2ax_sum,h2ax_mean)

    #sampling_rows
    
    
    
    rows.append([image, objects,group, h2ax_sum,h2ax_mean, positive])

        
        
        
sampling_summary = pd.DataFrame(rows, columns=['ImageNumber', 'Nuclei Count','Group', 'Parent_H2AX_Sum','Parent_H2AX_Mean','Positive Nuclei'])
sampling_summary
```

```python
sampling_summary.to_csv('samplingSummary.csv')
```

```python
#groups
df_groups = sampling_summary.groupby('Group')

df_controls = df_groups.get_group('Control')
df_PGVG = df_groups.get_group('PGVG')
df_Nic1 = df_groups.get_group('12.5mg')
df_Nic2 = df_groups.get_group('25mg')
df_Nic3 = df_groups.get_group('50mg')
df_CSE = df_groups.get_group('CSE')
```
